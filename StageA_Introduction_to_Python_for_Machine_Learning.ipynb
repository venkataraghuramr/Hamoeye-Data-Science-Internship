## Question 12 Select columns 'Y2017' AND 'Area'. Perform a group operation on 'Area'. Which of these areas had highest sum in 2017.

Ans - Nigeria

```python
question12 = df.groupby('Area')['Area','Y2017'].sum()
sortedDF = question12.sort_values('Y2017', ascending=False)
print(sortedDF)
```
